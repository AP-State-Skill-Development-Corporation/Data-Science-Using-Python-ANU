```python
url = 'https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Data%20Analysis/Advertising.csv'
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df = pd.read_csv(url, index_col = 0)
df.head()
```

### Distribution befor Standardization of data

```python
sns.kdeplot(df['TV'])
sns.kdeplot(df['radio'])
sns.kdeplot(df['newspaper'])
sns.kdeplot(df['sales'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
stdScale = std.fit_transform(df)
```

```python
stdScale
```

```python
df_scale = pd.DataFrame(stdScale, columns=df.columns)

df_scale.head()
```

```python
df_scale.mean().round(2)
```

```python
df_scale.std().round(2)
```

```python
sns.kdeplot(df_scale['TV'])
sns.kdeplot(df_scale['radio'])
sns.kdeplot(df_scale['newspaper'])
sns.kdeplot(df_scale['sales'])
```

```python
df.plot(kind = 'box')
```

```python
from sklearn.preprocessing import robust_scale
```

```python
rsc = robust_scale(df)
```

```python
df_scale = pd.DataFrame(rsc, columns = df.columns)
```

```python
sns.kdeplot(df_scale['TV'])
sns.kdeplot(df_scale['radio'])
sns.kdeplot(df_scale['newspaper'])
sns.kdeplot(df_scale['sales'])
```
